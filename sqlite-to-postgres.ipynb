## SQLite to Postgres

```python
import os
```

```python
sqlite_files = []
for root, dirs, files in os.walk("../spider/spider/database/"):
 for file in files:
    if file.endswith('.sqlite'):
       sqlite_files.append(os.path.join(root, file))
sqlite_files
```

```python
sqlite_names = [name.split('/')[4] for name in sqlite_files]
sqlite_names
```

```python
import subprocess
from tqdm.contrib import tzip

for db_name, sqlite_file in tzip(sqlite_names, sqlite_files):
    subprocess.run(["createdb", db_name])
    subprocess.run(["pgloader", sqlite_file, f'postgresql:///{db_name}'])
```
