## Postgres to Hasura

```python
import os
import requests

sqlite_files = []
for root, dirs, files in os.walk("../spider/spider/database/"):
 for file in files:
    if file.endswith('.sqlite'):
       sqlite_files.append(os.path.join(root, file))
sqlite_names = [name.split('/')[4] for name in sqlite_files]
sqlite_names
```

```python
from tqdm.notebook import tqdm
for db_name in tqdm(sqlite_names):
    x = requests.post('http://localhost:8080/v1/metadata', json={
      "type": "pg_drop_source",
      "args": {
        "name": db_name,
        "cascade": True        
       }
    })
```

```python
from tqdm.notebook import tqdm
for db_name in tqdm(sqlite_names):
    x = requests.post('http://localhost:8080/v1/metadata', json={
      "type": "pg_add_source",
      "args": {
        "name": db_name,
        "customization": {
          "root_fields": {
            "namespace": db_name,
            "prefix": db_name + "_"
          },
          "type_names": {
            "prefix": db_name + "_",
          },
        },
        "configuration": {
          "connection_info": {
            "database_url": "postgres://user@172.18.0.1/" + db_name,
            "pool_settings": {
              "retries": 1,
              "idle_timeout": 180,
              "max_connections": 50
            }
          }
        }
      }
    })
```
